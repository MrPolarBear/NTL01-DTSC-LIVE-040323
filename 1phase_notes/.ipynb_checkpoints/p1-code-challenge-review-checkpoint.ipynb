## Task 14: Create a new dictionary

Create a dictionary called `song_dictionary` which consists of each track name `string` as a key and a `tuple` of artists associated with each track as a value.

```python
# call_on_students(1)
```

```python
# Your code here:
song_dictionary = {}



for song in first_twenty_songs:
    artists_name_list = []
    artists = song['artists']
    for artist in artists:
        artists_name_list.append(artist['name'])
    #song_dictionary.update( {song['name']: tuple(artists_name_list)}  )
    song_dictionary[song['name']] = tuple(artists_name_list) 
    
song_dictionary
```
